```python
# add modules folder to Python's search path
from os import times
import sys
from pathlib import Path
from os.path import dirname, realpath
#script_dir = 'C:/Users/Nelissen/Documents/Github/NMA/experiments/rest/'
#module_dir = 'C:/Users/Nelissen/Documents/Github/NMA/'
script_dir = 'C:/Users/pinak/Documents/Github/NMA/experiments/rest/'
module_dir = 'C:/Users/pinak/Documents/Github/NMA/'
sys.path.insert(0, module_dir + '/modules')
print(module_dir)
print(script_dir)

import rest
from sklearn.decomposition import FastICA, PCA
import matplotlib.pyplot as plt
import numpy as np
import os

# a cursory look at the data
db_path = '{}/data/hcp_rest'.format(module_dir)
subject_id = 15
timeseries_index = 2
subject = rest.Individual(db_path, subject_id) 
timeseries = subject.get_timeseries(timeseries_index)
print(subject.get_regions())
print(timeseries.shape)                                                     #print
regions = subject.get_regions()
subject_folder = 'subject_{}'.format(subject_id)
if not os.path.isdir(subject_folder):
    os.mkdir(subject_folder)
#print(regions, regions.shape)


# PCA on the timeseries to estimate the number of components needed to explain 90% of variance

def do_pca(data):
    pca = PCA(n_components=200)
    
    # PCA on the data
    sources = pca.fit_transform(data.T)
    variance_data = np.cumsum(pca.explained_variance_ratio_)
    
    # PCA threshold line
    threshold =  np.full(200, 0.9) # threshold line at 90% variance explained
    
    # scree plot
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111)
    x=list(range(1, 201))
    ax.scatter(x, variance_data)
    ax.scatter(x, threshold)
    ax.set_xlabel('principal components')
    ax.set_ylabel('variance explained')
    
    diff = threshold - variance_data
    index = [i for i, x in enumerate(diff) if x > 0][-1]
    print(index)
    # save figure
    #plt.savefig('{}/scree_plot_ts_{}.png'.format(subject_folder, timeseries_index)) 
    #plt.close(fig)
    return

do_pca(timeseries)


```
